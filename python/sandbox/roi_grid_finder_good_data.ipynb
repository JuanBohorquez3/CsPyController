```python
# define gaussian with various degrees of freedom

def normal_gaussian(a,xy):
    return a*numpy.exp(-0.5*(np.sum(xy**2,axis=0)))

def elliptical_gaussian(a,w,xy):
    return normal_gaussian(a,xy/w)

def rotation(angle):
    return np.array([[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]])

def rotated_gaussian(a,w,rotation,xy):
    #xy is (2,len(x),len(y)) as such returned by np.indices((x,y))
    r = np.array([[np.cos(rotation), -np.sin(rotation)], [np.sin(rotation), np.cos(rotation)]])
    xy=np.tensordot(r,xy,axes=1)
    return elliptical_gaussian(a,w,xy)

def gaussian(a,w,rotation,xy0,xy):
    return rotated_gaussian(a,w,rotation,xy-xy0)

def g(xy,a,wx,wy,rotation,x0,y0):
    #useful for optimizer where parameters must be scalars
    return gaussian(a,numpy.array([[[wx]],[[wy]]]),rotation,numpy.array([[[x0]],[[y0]]]),xy)

def fit_rotation(data,xy,guess):
    error = lambda p: np.ravel(g(xy,*p)-data)
    p, success = leastsq(error, guess)
    return p
```

```python
# guess the top-left and bottom-right corners of the array as (row,column) = (top,left), (bottom,right)
top = 7
left = 21
bottom = 47
right = 58

#specify the size of the array
rows = 7
columns = 7

#find the width and height of the whole array
span = numpy.array([bottom-top,right-left])
#print 'span =',span
#find the proper diagonal angle for the array
proper_angle = numpy.arctan2(columns-1,rows-1)

# find the tilt angle deviation
angle = numpy.arctan2(span[1],span[0]) - proper_angle
#print 'angle =',angle

# find the spacing by using the diagonal distance
diagonal_distance = numpy.sqrt(numpy.sum(span**2))
diagonal_units = numpy.sqrt((rows-1)**2+(columns-1)**2)
spacing = diagonal_distance / diagonal_units

#guess the gaussian width
width = spacing/4

#find the noise level, and the height of the gaussians.  This could be improved using a fit to the histogram
amplitude = amax(image_sum)
blacklevel = amin(image_sum)

initial_guess = (top,left,spacing,angle,amplitude,width,width,0,blacklevel)
```

```python
# find the best fit to a square grid of gaussians of all the same height, equal spacing in x and y,
# allow for rotation of array, allow for different wx and wy, allow for uniform rotation of spots

def fitFunc(xy,x0,y0,spacing,grid_angle,amplitude,wx,wy,spot_angle,blacklevel):
    # sum up the contribution of each gaussian to the total
    xy0 = np.array([[[x0]], [[y0]]])
    width = np.array([[[wx]], [[wy]]])
    spots = []
    for r in xrange(rows):
        for c in xrange(columns):
            xy0i = xy0 + np.tensordot(rotation(grid_angle),np.array([[[r]],[[c]]]),axes=1)*spacing
            spots.append(gaussian(amplitude, width, spot_angle, xy0i, xy))
    out = numpy.sum(spots, axis=0)+blacklevel
    return out.ravel()

#create the (x,y) values for each point on the image
xy = np.indices(image_sum.shape)

#use the image_sum as our real data
y = image_sum.ravel()

fitParams, fitCovariances = curve_fit(fitFunc, xy, y, p0=initial_guess)
print ' initial guess: top {}, left {}, spacing {}, angle {}, amplitude {}, width {}, blacklevel {}\n'.format(*initial_guess)
print ' fit coefficients: top {}, left {}, spacing {}, angle {}, amplitude {}, width {}, blacklevel {}\n'.format(*fitParams)
print ' Covariance matrix:\n', fitCovariances

gs=GridSpec(1,5)
fig=figure(figsize=(50,10))

# plot the original data
ax=fig.add_subplot(gs[0,0])
ax.matshow(numpy.sum(images, axis=(0,1)))
ax.set_title('raw data')

# plot the ICA cleaned up version
ax=fig.add_subplot(gs[0,1])
p1=ax.matshow(image_sum)
#fig.colorbar(p1)
ax.set_title('cleaned up with ICA')

#plot the guess
ax=fig.add_subplot(gs[0,2])
ax.matshow(fitFunc(xy,*initial_guess).reshape(image_sum.shape[0],image_sum.shape[1]),vmin=amin(image_sum),vmax=amax(image_sum))
ax.set_title('guess')

#plot the fit
ax=fig.add_subplot(gs[0,3])
ax.matshow(fitFunc(xy,*fitParams).reshape(image_sum.shape[0],image_sum.shape[1]),vmin=amin(image_sum),vmax=amax(image_sum))
ax.set_title('fit')

#plot the 1 sigma gaussian ellipses
x0,y0,spacing,grid_angle,amplitude,wx,wy,spot_angle,blacklevel = fitParams
xy0 = np.array([[[x0]], [[y0]]])
width = np.array([[[wx]], [[wy]]])
ax=fig.add_subplot(gs[0,4])
ax.matshow(image_sum)
ax.set_title('1 sigma contour on data')
for r in xrange(rows):
    for c in xrange(columns):
        xy0i = xy0 + np.tensordot(rotation(grid_angle),np.array([[[r]],[[c]]]),axes=1)*spacing
        patch = Ellipse(xy0i[::-1], width=2*wx, height=2*wy, angle=spot_angle, edgecolor='white', facecolor='none', lw=1)
        ax.add_patch(patch)
        ax.text(xy0i[1],xy0i[0],str(rows*r+c))
```

```python
revised_guess = numpy.zeros(1+6*rows*columns)
revised_guess[0] = blacklevel
for r in xrange(rows):
    for c in xrange(columns):
        xy0i = xy0 + np.tensordot(rotation(grid_angle),np.array([[[r]],[[c]]]),axes=1)*spacing
        i = 1+6*(r*columns+c)
        revised_guess[i:i+6] = [amplitude,xy0i[0,0,0],xy0i[1,0,0],wx,wy,spot_angle]
```

```python
# now fit again with variable positions and width on each one, but using the regular grid as an initial guess

def fitFunc(xy,params):
    # sum up the contribution of each gaussian to the total
    blacklevel = params[0]
    spots = []
    for i in xrange(rows*columns):
        x0 = params[1+5*i]
        y0 = params[2+5*i]
        wx = params[3+5*i]
        wy = params[4+5*i]
        spot_angle = params[5+5*i]
        xy0i = np.array([[[x0]], [[y0]]])
        width = np.array([[[wx]], [[wy]]])
        spots.append(gaussian(amplitude, width, spot_angle, xy0i, xy))
    out = numpy.sum(spots, axis=0)+blacklevel
    return out

def errorfunc(params):
    return np.ravel(fitFunc(xy,params)-image_sum)

fitParams, success = leastsq(errorfunc, revised_guess)
print success
```
