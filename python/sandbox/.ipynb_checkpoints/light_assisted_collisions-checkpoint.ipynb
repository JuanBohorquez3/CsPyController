```python
from __future__ import division
from sympy import sqrt
from sympy.physics.wigner import wigner_3j, wigner_6j
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

def V_DD_i(phi,parity,I,J,G1,F1,mF1,G,F,mF):
    return float(parity*(-1)**(2*I+G+G1+F+F1+mF+mF1)*(J+1/2)
            *sqrt((2*G+1)*(2*G1+1)*(2*F+1)*(2*F1+1))
            *wigner_6j(F1,1,G,1/2,I,J)*wigner_6j(F,1,G1,1/2,I,J)
            *(2*wigner_3j(F1,G,1,-mF1,phi-mF,0)*wigner_3j(F,G1,1,-mF,phi-mF1,0)
              -wigner_3j(F1,G,1,-mF1,phi-mF,1)*wigner_3j(F,G1,1,-mF,phi-mF1,1)
              -wigner_3j(F1,G,1,-mF1,phi-mF,-1)*wigner_3j(F,G1,1,-mF,phi-mF1,-1))
            )

def V_DD_state(phi,parity,I,J,state1,state2):
    return V_DD_i(phi,parity,I,J,state1[0],state1[1],state1[2],state2[0],state2[1],state2[2])

#V_DD = np.vectorize(V_DD_state)

def molecular_state(atom,phi,parity):
    # phi is the on-axis angular momentum
    # parity is +1 for even (u) states and -1 for odd (g) states
    
    I=atom['I']
    # start in the highest S_1/2+S_1/2 state
    G_list = np.arange(np.abs(I-1/2), I+1/2+1)  # the +1 makes the range inclusive
    G = G_list[-1]
    F = G

    # always go up to an S_1/2+P_3/2 state
    J = 3/2
    F_list = np.arange(np.abs(I-J), I+J+1)  # the +1 makes the range inclusive

    # figure out what sets of G,F,mF are possible
    states = []
    for G1 in G_list:
        for F1 in F_list:
            for mF1 in np.arange(-F1, F1+1):  # the +1 makes the range inclusive
                mG1 = phi-mF1
                if -G1<=mG1<=G1:
                    states.append([G1,F1,mF1])
    
    # evaluate the Hamiltonian
    n=len(states)
    H = np.zeros((n,n))
    for i in xrange(n):
        for j in xrange(n):
            H[i,j] = V_DD_state(phi,parity,I,J,states[i],states[j])
            # subtract the hyperfine splitting
            lowest_S_state=np.abs
            H[i,j] += (statesG1-G)*atom['hyperfine'] + atom['hyperfine_P'][F-F_list[0]]

    for i in state
    # evaluate the matrix element for every combination
    #    a = V_DD(phi,parity,G1,J,F1,mF1,G,F,mF,I)
    
    #    for i in states:
    #        b=E(phi,parity,i[0],J,i[1],i[2],G,F,mF,I)
    #        print b
    #        c=s1*b
    #        plt.plot(s,c);

        # scan over a range of interatomic spacings
    # s = C_3/(h*R**3) in GHz gives result in E/h, GHz
    s = np.linspace(20,0,10)
    s1=s.reshape((len(s),1))

    
def E(phi,parity,G1,J,F1,mF1,G,F,mF,I):
    # find each matrix element
    a = V_DD(phi,parity,G1,J,F1,mF1,G,F,mF,I)
    
    # calculate the eigenvalues to get the energies
    b = np.linalg.eigvals(a)
    
    # subtract the hyperfine splitting
    lowest_P_state=np.abs(I-J)
    return b + (G1-G)*atom['hyperfine'] + atom['hyperfine_P'][F-lowest_P_state]
```
