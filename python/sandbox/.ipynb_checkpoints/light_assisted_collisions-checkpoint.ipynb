Recreating plots from Laser Physics vol. 4 no. 5 1995 pp.1085-1092, T. Walker and D. Pritchard

```python
from __future__ import division
from sympy import sqrt, S
from sympy.physics.wigner import wigner_3j, wigner_6j
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

#atomic data.  Hyperfine levels relative to highest F state, in GHz
Rb85 = {'hyperfine':{'S':3.036, 'P':{1/2:[],3/2:[-.203,-.184,-.121,0]}}, 'I':5/2}
Rb87 = {'hyperfine':{'S':6.835, 'P':{1/2:[],3/2:[-.496,-.424,-.267,0]}}, 'I':3/2}
Cs133 = {'hyperfine':{'S':9.192631770, 'P':{1/2:[-1.167680,0],3/2:[-.604,-.453,-.252,0]}}, 'I':7/2}

# calculate an individual matrix element for the dipole-dipole interaction
def V_DD_i(phi,parity,I,J,G1,F1,mF1,G,F,mF):
    return float(parity*(-1)**(2*I+G+G1+F+F1+mF+mF1)*(J+1/2)
            *sqrt((2*G+1)*(2*G1+1)*(2*F+1)*(2*F1+1))
            *wigner_6j(F1,1,G,S(1)/2,I,J)*wigner_6j(F,1,G1,S(1)/2,I,J)  # Using a symbolic S(1)/2 prevents an error on the first run, I don't know why.
            *(2*wigner_3j(F1,G,1,-mF1,phi-mF,0)*wigner_3j(F,G1,1,-mF,phi-mF1,0)
              -wigner_3j(F1,G,1,-mF1,phi-mF,1)*wigner_3j(F,G1,1,-mF,phi-mF1,1)
              -wigner_3j(F1,G,1,-mF1,phi-mF,-1)*wigner_3j(F,G1,1,-mF,phi-mF1,-1))
            )

#calculate an individual matrix element for the dipole-dipole interaction
#accepts two state tuples of (G,F,mF) instead of individual G,F,mF parameters
def V_DD_state(phi,parity,I,J,state1,state2):
    return V_DD_i(phi,parity,I,J,state1[0],state1[1],state1[2],state2[0],state2[1],state2[2])

# find the potentials for all possible angular momenta phi for a given species
def all_molecular_states(title,atom,J):
    phimax = int((atom['I']+1/2)+(atom['I']+J))
    phis = range(0,phimax+1)  # the +1 makes the range inclusive
    fig, ax = plt.subplots(2, 2*len(phis), figsize=(5*len(phis),10))
    for phi in phis:
        print 'phi =',phi
        molecular_state('{} {}g'.format(title,phi),atom,phi,+1,J,ax[0,2*phi],ax[1,2*phi])
        molecular_state('{} {}u'.format(title,phi),atom,phi,-1,J,ax[0,2*phi+1],ax[1,2*phi+1])

# find the potential for a specific species, angular momentum phi, and parity
def molecular_state(title,atom,phi,parity,J,ax1,ax2):
    # phi is the on-axis angular momentum
    # parity is +1 for even (g) states and -1 for odd (u) states
    
    # scan over a range of interatomic spacings
    # s = C_3/(h*R**3) in GHz gives result in E/h, GHz
    s1 = np.linspace(20,0,100)
    s2 = np.linspace(2,0,100)

    I=atom['I']
    
    # start in the highest S_1/2+S_1/2 state
    G_list = np.arange(np.abs(I-1/2), I+1/2+1)  # the +1 makes the range inclusive
    
    # always go up to an S_1/2+P_j state
    F_list = np.arange(np.abs(I-J), I+J+1)  # the +1 makes the range inclusive

    # figure out what sets of G,F,mF are possible
    states = []
    for G1 in G_list:
        for F1 in F_list:
            for mF1 in np.arange(-F1, F1+1):  # the +1 makes the range inclusive
                mG1 = phi-mF1
                if -G1<=mG1<=G1:
                    states.append([G1,F1,mF1])
    n=len(states)
                
    if n>0:
        print 'states:\n',states

        # evaluate the electrostatic dipole-dipole interaction
        V_DD = np.empty((n,n))
        #print 'elements'
        for i in xrange(n):
            for j in xrange(n):
                #print states[i],states[j]
                V_DD[i,j] = V_DD_state(phi,parity,I,J,states[i],states[j])

        # create a diagonal matrix with the hyperfine offsets
        H0 = np.zeros((n,n))
        for i in xrange(n):
            try:
                H0[i,i] = (states[i][0]-G_list[-1])*atom['hyperfine']['S'] + atom['hyperfine']['P'][J][int(states[i][1]-F_list[0])]
            except IndexError:
                print 'state', states[i]
                print 'G_list', G_list
                print 'F_list', F_list
                print 'index',int(states[i][1]-F_list[0])
                return
        
        # find the energies over a large range
        E = np.empty((len(s1),n))
        for i,si in enumerate(s1):

                # create the hamiltonian for this spacing
                H = H0 + si*V_DD

                # calculate the eigenvalues to get the energies
                E[i] = np.sort(np.linalg.eigvals(H))

        # plot
        ax1.plot(s1,E)
        ax1.set_title(title)
        ax1.set_ylim(-20,20)
        ax1.set_xlim(20,0)

        # find the energies near infinite separation
        E = np.empty((len(s2),n))
        for i,si in enumerate(s2):

                # create the hamiltonian for this spacing
                H = H0 + si*V_DD

                # calculate the eigenvalues to get the energies
                E[i] = np.sort(np.linalg.eigvals(H))

        # plot
        ax2.plot(s2,E)
        ax2.set_ylim(-1,.2)
        ax2.set_xlim(2,0)
```

```python
fig, ax = plt.subplots(2, 4, figsize=(20,10))
J=3/2

# Fig. 4. 85Rb 3u
molecular_state('85Rb 3u',Rb85,3,-1,J,ax[0,0],ax[1,0])

# Fig. 5. 87Rb 4g
molecular_state('87Rb 4g',Rb87,4,+1,J,ax[0,1],ax[1,1])

# Fig. 6. 87Rb 4u
molecular_state('87Rb 4u',Rb87,4,-1,J,ax[0,2],ax[1,2])

# Fig. 7. 87 Rb 3g
molecular_state('87Rb 3g',Rb87,3,+1,J,ax[0,3],ax[1,3])
```

```python
# Cesium
fig, ax = plt.subplots(2, 1, figsize=(5,10))
J=3/2
molecular_state('133Cs 3g',Cs133,3,+1,J,ax[0],ax[1])
```

```python
# Cesium P_1/2
J=1/2
all_molecular_states('133Cs',Cs133,J)
```
