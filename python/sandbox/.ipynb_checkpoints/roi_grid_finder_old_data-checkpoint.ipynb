```python
# import images from hdf5 file
import h5py
f = h5py.File(r'D:\users\Public\AQuA data\2014 data\2014_05_20\2014_05_20_07_01_23_retention\results.hdf5','r')
images = numpy.array([[m['data/Hamamatsu/shots/0'].value for m in i['measurements'].itervalues()] for i in f['iterations'].itervalues()])
print images.shape
f.close()
```

```python
# find the image summation
image_sum = numpy.sum(images, axis=(0,1))
matshow(image_sum)
```

```python
# clean up using independent component analysis

X = images.reshape(images.shape[1],images.shape[2]*images.shape[3]).astype(float)
#X -= numpy.mean(X,axis=0)
#X /= X.std(axis=0)  # Standardize data

# Compute ICA
ica = FastICA(n_components=49,max_iter=2000)
ica.fit(X)
A_ica = ica.components_  # Get estimated mixing matrix

image_sum=numpy.sum(numpy.abs(A_ica),axis=0).reshape(50,103)
matshow(image_sum)

#pca = PCA(n_components=49)
#pca.fit(X)
#A_pca = pca.components_
#matshow(numpy.sum(numpy.abs(A_pca),axis=0).reshape(50,103))
```

```python
# guess the top-left and bottom-right corners of the array as (row,column) = (top,left), (bottom,right)
top = 7
left = 21
bottom = 47
right = 58

#specify the size of the array
rows = 7
columns = 7

#guess the gaussian width
width = 1

#find the width and height of the whole array
span = numpy.array([bottom-top,right-left])
#print 'span =',span
#find the proper diagonal angle for the array
proper_angle = numpy.arctan2(rows-1,columns-1)

# find the tilt angle deviation
angle = numpy.arctan2(span[0],span[1]) - proper_angle
#print 'angle =',angle

# find the spacing by using the diagonal distance
diagonal_distance = numpy.sqrt(numpy.sum(span**2))
diagonal_units = numpy.sqrt((rows-1)**2+(columns-1)**2)
spacing = diagonal_distance / diagonal_units

#find the noise level, and the height of the gaussians.  This could be improved using a fit to the histogram
amplitude = amax(image_sum)
blacklevel = amin(image_sum)

initial_guess = (top,left,spacing,angle,amplitude,width,blacklevel)
```

```python
# find the best fit to a square grid of gaussians of all the same height and width (same in x and y), equal spacing in x and y
# allow for rotation

import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def gaussian(a,w,x,y):
    return a*numpy.exp(-0.5*(x**2+y**2)/w**2)

row_array, column_array = numpy.meshgrid(xrange(rows),xrange(columns))

def fitFunc((x,y),top,left,spacing,angle,amplitude,width,blacklevel):
    # sum up the contribution of each gaussian to the total
    spots = []
    for r in xrange(rows):
        for c in xrange(columns):
            xi = x-(top+r*spacing*numpy.cos(angle)+c*spacing*numpy.sin(angle))
            yi = y-(left-r*spacing*numpy.sin(angle)+c*spacing*numpy.cos(angle))
            spots.append(gaussian(amplitude, width, xi, yi))
    
    out = numpy.sum(spots, axis=0)+blacklevel
    return out.ravel()

#create the (x,y) values for each point on the image
height, width = image_sum.shape
x1, x2 = numpy.meshgrid(xrange(height),xrange(width))

#use the image_sum as our real data
y = image_sum.ravel()

fitParams, fitCovariances = curve_fit(fitFunc, (x1.T, x2.T), y, p0=initial_guess)
print ' initial guess: top {}, left {}, spacing {}, angle {}, amplitude {}, width {}, blacklevel {}\n'.format(*initial_guess)
print ' fit coefficients: top {}, left {}, spacing {}, angle {}, amplitude {}, width {}, blacklevel {}\n'.format(*fitParams)
print ' Covariance matrix:\n', fitCovariances

gs=GridSpec(1,3)
fig=figure(figsize=(20,10))

# plot the data
ax=fig.add_subplot(gs[0,0])
p1=ax.matshow(image_sum)
#fig.colorbar(p1)
ax.set_title('data')

#plot the guess
ax=fig.add_subplot(gs[0,1])
ax.matshow(fitFunc((x1.T,x2.T),*initial_guess).reshape(image_sum.shape[0],image_sum.shape[1]),vmin=amin(image_sum),vmax=amax(image_sum))
ax.set_title('guess')

#plot the fit
ax=fig.add_subplot(gs[0,2])
ax.matshow(fitFunc((x1.T,x2.T),*fitParams).reshape(image_sum.shape[0],image_sum.shape[1]),vmin=amin(image_sum),vmax=amax(image_sum))
ax.set_title('fit')
```

```python
# Use the regular grid as a starting point for regions of interest that allow more degrees of freedom

# for each spot, use only data in a square at xi +- w and yi +- w
# fit the data in that square to a gaussian to get an updated xi, yi, w

z = image_sum.astype(float)
n = rows*columns

def aargmax(z):
    w,h=shape(z)
    i=argmax(z)
    return int(i/h), i%h

def gaussian2D(r,a,w):
    #print 'r=',r
    #print 'a=',a
    #print 'w=',w
    g = a/(2*pi*w**2)*exp(-0.5*r**2/w**2)
    #print 'g=',g
    g[isnan(g)] = 0 #eliminate bad elements
    return g

top,left,spacing,angle,amplitude,sigma,blacklevel = fitParams

x_tot, y_tot = z.shape

gs=numpy.zeros((n,x_tot,y_tot))
meanxs=numpy.zeros(n)
meanys=numpy.zeros(n)
widths=numpy.zeros(n)
amplitudes=numpy.zeros(n)
errors=numpy.zeros(n)

datalump = numpy.zeros_like(z)
datalist=[]
i=0
for row in xrange(rows):
    for column in xrange(columns):
        
        #use the regular grid fit as a starting point
        x0 = (top+row*spacing*numpy.cos(angle)+column*spacing*numpy.sin(angle))
        y0 = (left-row*spacing*numpy.sin(angle)+column*spacing*numpy.cos(angle))
        
        data = numpy.zeros_like(z)
        
        x1=int(numpy.ceil(x0-spacing/2.0))
        if x1<0: x1=0
        elif x1>x_tot: x1=x_tot
        
        x2=int(numpy.floor(x0+spacing/2.0+1))
        if x2<0: x2=0
        elif x2>x_tot: x2=x_tot
        
        y1=int(numpy.ceil(y0-spacing/2.0))
        if y1<0: y1=0
        elif y1>y_tot: y1=y_tot
        
        y2=int(numpy.floor(y0+spacing/2.0+1))
        if y2<0: y2=0
        elif y2>y_tot: y2=y_tot
        
        #copy the data for the active range only
        data[x1:x2,y1:y2] = z[x1:x2,y1:y2]-blacklevel
        datalump += data
        datalist.append(data)
        
        #an array of x,y positions
        xy=numpy.indices((x_tot,y_tot))
        x=xy[0]
        y=xy[1]
        
        #fit a gaussian with blacklevel offset
        meanx=numpy.sum(x*data)/numpy.sum(data)
        #print 'meanx=',meanx
        meany=numpy.sum(y*data)/numpy.sum(data)
        #print 'meany=',meany
        r=numpy.sqrt((x-meanx)**2+(y-meany)**2) #an array of distances from the mean
        #matshow(r)
        width=numpy.sqrt(numpy.abs(numpy.sum((r**2)*data)/numpy.sum(data))) #initial fit assumes azimuthal symmetry
        #print 'width=',width
        amplitude=numpy.sum(data) #area under gaussian is 1, so scale by total volume (i.e. the sum of z)
        #print 'amp=',amplitude
        g=gaussian2D(r,amplitude,width)
        error=numpy.sum(numpy.abs(data-g))
        #store the fit
        gs[i]=g
        meanxs[i]=meanx
        meanys[i]=meany
        widths[i]=width
        amplitudes[i]=amplitude
        errors[i]=error
        i+=1

matshow(datalump)
```

```python
# Use a gaussian multiplier before doing the fit

# for each spot, use only data in a square at xi +- w and yi +- w
# fit the data in that square to a gaussian to get an updated xi, yi, w

z = image_sum.astype(float)
n = rows*columns

def aargmax(z):
    w,h=shape(z)
    i=argmax(z)
    return int(i/h), i%h

def gaussian2D(r,a,w):
    #print 'r=',r
    #print 'a=',a
    #print 'w=',w
    g = a/(2*pi*w**2)*exp(-0.5*r**2/w**2)
    #print 'g=',g
    g[isnan(g)] = 0 #eliminate bad elements
    return g

top,left,spacing,angle,amplitude,sigma,blacklevel = fitParams

x_tot, y_tot = z.shape

gs=numpy.zeros((n,x_tot,y_tot))
meanxs=numpy.zeros(n)
meanys=numpy.zeros(n)
widths=numpy.zeros(n)
amplitudes=numpy.zeros(n)
errors = numpy.zeros(n)

datalump = numpy.zeros_like(z)
datalist=[]
i=0
for row in xrange(rows):
    for column in xrange(columns):
        
        # use the regular grid fit as a starting point
        x0 = (top+row*spacing*numpy.cos(angle)+column*spacing*numpy.sin(angle))
        y0 = (left-row*spacing*numpy.sin(angle)+column*spacing*numpy.cos(angle))
        
        # an array of x,y positions
        xy=numpy.indices((x_tot,y_tot))
        x=xy[0]
        y=xy[1]
        
        # create a gaussian centered at x0, y0
        multiplier = gaussian(1/(2*numpy.pi*sigma**2),sigma,x-x0,y-y0)

        #use the gaussian to mask the data
        data = multiplier * (z-blacklevel)
        
        #copy the data for the active range only
        datalump += data
        datalist.append(data)

        #fit a gaussian with blacklevel offset
        meanx=numpy.sum(x*data)/numpy.sum(data)
        #print 'meanx=',meanx
        meany=numpy.sum(y*data)/numpy.sum(data)
        #print 'meany=',meany
        r=numpy.sqrt((x-meanx)**2+(y-meany)**2) #an array of distances from the mean
        #matshow(r)
        width=numpy.sqrt(numpy.abs(numpy.sum((r**2)*data)/numpy.sum(data))) #initial fit assumes azimuthal symmetry
        #print 'width=',width
        amplitude=numpy.sum(data) #area under gaussian is 1, so scale by total volume (i.e. the sum of z)
        #print 'amp=',amplitude
        g=gaussian2D(r,amplitude,width)
        error=numpy.sum(numpy.abs(data-g))
        #store the fit
        gs[i]=g
        meanxs[i]=meanx
        meanys[i]=meany
        widths[i]=width
        amplitudes[i]=amplitude
        errors[i]=error
        i+=1
matshow(datalump)
```
