```python
import os,datetime,numpy
import matplotlib.pyplot as plt
from matplotlib.dates import AutoDateLocator,AutoDateFormatter

#useful abbreviations
min=60
hour=60*min
day=24*hour

filenames=['topS','topN','bottomS','bottomN','frontS','frontN']
columns=['set point','main sensor','cold plate','extra sensor','controller','current','voltage','power']

def str2time(timeString):
    return datetime.datetime.strptime(timeString,'%Y/%m/%d %H:%M:%S')

vectorGetTime=numpy.vectorize(str2time)

def time2str(time):
    return datetime.datetime.fromtimestamp(time).strftime('%Y/%m/%d %H:%M:%S')

def plotTemp(t,path):
    
    #convert time range to float time
    cutoffTime=datetime.datetime.now()-t
    
    data={}
    timeStrings={}
    times={}
    cutoff={}
    for i in filenames:
        with open(os.path.join(path,i+'.txt'), 'r') as f:
            textIn=np.loadtxt(f,delimiter='\t',dtype='S19,f4,f4,f4,f4,f4,f4,f4,f4',skiprows=1,unpack=True)
            timeStrings[i]=textIn[0]
            data[i]=textIn[1:]
        
        #convert string dates
        times[i]=vectorGetTime(timeStrings[i])
        
        #find first date after cutoff date
        cutoff[i]=numpy.argmax(times[i]>=cutoffTime)
        
        if times[i][cutoff[i]]>=cutoffTime:
            #we found a good result
            #drop unneeded times now, so we only have to do this once
            times[i]=times[i][cutoff[i]:]
        else:
            #argmax did not find any good results, and so returned the first element.  This is the opposite of what we want
            print 'no data found in that time range'
            return
    
    #plot data
    fig=plt.figure(figsize=(20,40))
    linelist=[]
    for i,c in enumerate(columns):
        ax=fig.add_subplot(len(columns),1,i+1)
        ax.set_ylabel(c)
        for f in filenames:
            p=plt.plot(times[f],data[f][i][cutoff[f]:],label=f)
        #legend only on top plot
        if i==0:
            ax.legend(loc='upper center',ncol=len(columns),bbox_to_anchor=(0.5, 1.2), fancybox=True)
    fig.autofmt_xdate()
```
